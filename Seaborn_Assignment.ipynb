```python
# Code here and leave the next cell empty, so that you don't lose the plot!
# Load datasets
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
# Load the 'tips' dataset from seaborn
tips = sns.load_dataset("tips")

# Create a boxplot using seaborn
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="tip", hue="sex", data=tips, palette="Set2")

# Display the plot
plt.show()
```

**Recreate the distplot given below. Use titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
# JointGrid: Total Bill vs Tip
g = sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")
g.set_axis_labels("Total Bill", "Tip")
plt.show()


```

**Recreate the boxplot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
# Create the histogram for fare
sns.histplot(titanic['fare'], kde=False, color="green", bins=20)

# Set labels
plt.xlabel('Fare')
plt.ylabel('Count')

# Display the plot
plt.show()

```

**Recreate the plot below using titanic dataset. Hint: pay attention to x and y labels.** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!

# Create the boxplot
sns.boxplot(x="class", y="age", hue="alive", data=titanic, palette="Set2")

# Set labels
plt.xlabel('Class')
plt.ylabel('Age')

# Display the plot
plt.show()
```

**Recreate the plot below. Use titanic dataset. ** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
# JointGrid: Fare vs Age
g = sns.jointplot(x="fare", y="age", data=titanic, kind="scatter")
g.set_axis_labels("Fare", "Age")
plt.show()


```

**Can you plot how many male and female were on titanic? Separate them based on their class.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!


# Create the countplot
sns.countplot(x="sex", hue="class", data=titanic, palette="Set1")

# Set labels
plt.xlabel('Sex')
plt.ylabel('Count')

# Display the plot
plt.show()



```

**Recreate the plot below. Use titanic dataset. Hint: Matrix plots** 

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
# Matrix plot: Heatmap of Correlation
corr = titanic.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()


```
