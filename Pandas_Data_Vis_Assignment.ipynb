```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1)
df = pd.DataFrame(np.random.rand(1000,4),columns=['a', 'b', 'c','d'])
#df = pd.DataFrame(np.random.randn(1000,4),columns=['a', 'b', 'c','d'])
```

**You have created a dataset `df`. Plot `c vs d` as a scatter plot to recreate the plot below. Try to match the color and size of the data points. You should be able to match the size of the figure as well!**

```python
# Code goes here!
# Scatter plot for c vs d
plt.figure(figsize=(8, 3))
plt.scatter(df['c'], df['d'], color='g', s=20)
plt.title('Scatter Plot of c vs d')
plt.xlabel('c')
plt.ylabel('d')
plt.show()
```

** Create a histogram of the 'd' column.**

```python
# Code goes here!
# Histogram of the 'd' column
plt.figure(figsize=(5, 3))
plt.hist(df['d'], bins=10, color='#4288c2')
plt.title('Histogram of column d')
plt.xlabel('d')
plt.ylabel('Frequency')
plt.show()
```

** Change style sheets to set for 'bmh' and add more bins (e.g. 50) to the above plot.**

```python
# Code goes here!
# Changing style sheets to 'bmh' and increasing the number of bins
plt.style.use('bmh')
plt.figure(figsize=(5, 3))
plt.hist(df['d'], bins=50, color='#abcae4')
plt.title('Histogram of column d with bmh style')
plt.xlabel('d')
plt.ylabel('Frequency')
plt.show()
```

**Create a scatter plot for a vs b. How to use the s parameter to indicate the size of the data point based of col d?**<br>

```python
# Code goes here!
plt.figure(figsize=(5,3))
plt.scatter(df['a'], df['b'], s=df['d']*100, alpha=0.5, color='g')
plt.title('Scatter Plot of a vs b with size based on d')
plt.xlabel('a')
plt.ylabel('b')
plt.show()
```

**Recreate the plot below**

```python
# Code goes here!


# Create a hexbin plot for 'a' vs 'b'
plt.figure(figsize=(5,3))
hb = plt.hexbin(df['a'], df['b'], gridsize=17, cmap='coolwarm')

# Add color bar
cb = plt.colorbar(hb, label='count')

# Add labels
plt.xlabel('a')
plt.ylabel('b')
plt.show()
```

**Create a boxplot comparing all columns in the data.**<br>
Hint: use `color = dict(boxes='DarkGreen', whiskers='DarkOrange',medians='DarkBlue', caps='Gray')` and `sym = 'r+'` 

```python
# Define the color dictionary as per your request
color = dict(boxes='green', whiskers='orange', medians='blue', caps='#808080')

# Create the boxplot
plt.figure(figsize=(5,3))
df.boxplot(color=color, sym='r+')

# Show the plot
plt.title('Boxplot comparing all columns')
plt.show()
```
