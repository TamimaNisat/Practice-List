```python
# We need data to work with, lets create data first!
import numpy as np
x = np.arange(0,10)
y = x*2
z = x**2
```

**1. Instructions:**
* ** Use <code>plt.figure()</code> and create a figure object, say fig.**
* ** Add an axis to the figure canvas (`fig`), `[left, bottom, width, height] is [0,0,1,1]`. <i>Hint: <code>add_axes()</code><i>**
* ** Plot the data x,y on the axes**
* ** Set the labels x and y**
* ** Set the title**
* ** Create legends as shown in the plot**

Match the output below: 

```python
#Code here
# Create figure and axis
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])  # [left, bottom, width, height]

# Plot the data
ax.plot(x, y, label='y = 2*x')

# Set labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Title')

# Add legend
ax.legend()

# Display the plot
plt.show()

```

**2. Instructions**<br>
* ** Create a figure object**
* ** Put two axes on the figure at [0,0,1,1] and [0.1,0.6,0.2,0.2]**
* ** Plot data (x,z) on main axes**
* ** Plot data (x,z) on the small axes and set its x limit b/w 1, 2, y limit 0, 5**
* ** Replicate the figure to axes, titles, legends etc**

```python
#Code here
# Create figure
fig = plt.figure()

# Add main axes and smaller axes
main_ax = fig.add_axes([0, 0, 1, 1])
small_ax = fig.add_axes([0.1, 0.6, 0.2, 0.2])

# Plot on main axes
main_ax.plot(x, x**2, label='y = x^2')
main_ax.set_xlabel('x_main')
main_ax.set_ylabel('y_main')
main_ax.set_title('Title_main')

# Plot on small axes with limits
small_ax.plot(x, x**2, label='y = x^2')
small_ax.set_xlim([1, 2])
small_ax.set_ylim([0, 5])
small_ax.set_title('inset')
small_ax.set_xlabel('x_inset')
small_ax.set_ylabel('y_inset')


# Show plot
plt.show()
```

**3. Replicate the output using <code>subplots()</code> and for loop** -- Note, the data is x and y. 

```python
#Code here
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

for ax in axes:
    ax.plot(x, y, label='y = 2x')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.legend()

plt.show()
```

**5. plot `(x,y)`, `(x,z)` and `(x,y)` on the above empty canvas to replicate the plot given below.**

```python
#Code here
# Data for the plots
x = np.arange(0, 10, 1)
y = x
z = x**2

# Creating subplots
fig, axs = plt.subplots(3)

# First subplot (blue dashed line, plot (x, y))
axs[0].plot(x, y, 'b--')
axs[0].set_title('Plot (x, y)')

# Second subplot (red solid line, plot (x, z))
axs[1].plot(x, z, 'r-')
axs[1].set_title('Plot (x, z)')

# Third subplot (green solid line, plot (x, y))
axs[2].plot(x, y, 'g-')
axs[2].set_title('Plot (x, y)')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()

```

**6. Resize the above plots to `figsize = (3,4)`**

```python
#Code here

# Creating subplots with figsize (3, 4)
fig, axs = plt.subplots(3, figsize=(3, 4))

# First subplot (blue dashed line, plot (x, y))
axs[0].plot(x, y, 'b--')
axs[0].set_title('Plot (x, y)')

# Second subplot (red solid line, plot (x, z))
axs[1].plot(x, z, 'r-')
axs[1].set_title('Plot (x, z)')

# Third subplot (green solid line, plot (x, y))
axs[2].plot(x, y, 'g-')
axs[2].set_title('Plot (x, y)')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()

```
