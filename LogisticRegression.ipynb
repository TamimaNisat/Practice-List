```python
import pandas as pd
from matplotlib import pyplot as plt
```

**load the csv**

```python
#code here

# Load CSV file into DataFrame
df = pd.read_csv('titanic_full.csv')

# Set option to display all rows
pd.set_option('display.max_rows', None)

# Set option to display all columns
pd.set_option('display.max_columns', None)
print(df)

```

**Check the null value**

```python
#code here
# Check for null values
null_values = df.isnull().sum()
print(null_values)
```

**Plot the data**

```python
#code here


# Plotting 'Fare' against 'Survived' using a scatter plot
plt.scatter(df['Fare'], df['Survived'], color='red')

# Adding labels and title
plt.xlabel('Fare')
plt.ylabel('Survived')

# Display the plot
plt.show()
```

**Separate the Features a target in x and y**

```python
#code here  
# Step 5: Separate features (X) and target (y)
X = df[['Pclass', 'Fare']]  
y = df['Survived']         
```

**Split the train and test data**

```python
#code here
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=50)
```

**Make the model Binary logistic regression model**

```python
#code here
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

**Check the model score**

```python
#code here
model.fit(X_train, y_train)

# Check the model accuracy on the test data
score = model.score(X_test, y_test)
print(f"Model Accuracy: {score:.4f}")
```
